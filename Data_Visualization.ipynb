```python
import pandas as pd

df = pd.read_csv(
    "netflix_titles.csv",
    encoding="latin1",
    engine="python",
    on_bad_lines="skip"
)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.countplot(y='rating', data=df)
plt.title("Ratings Distribution on Netflix")
plt.show()
```

```python
movies = df[df['type'] == 'Movie'].copy()
movies['duration_int'] = movies['duration'].str.replace(' min','', regex=False).astype(float)

plt.figure()
sns.histplot(movies['duration_int'], bins=30)
plt.title("Movie Duration Distribution")
plt.xlabel("Minutes")
plt.show()
```
