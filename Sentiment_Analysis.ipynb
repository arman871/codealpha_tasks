```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

data = {
    'review': [
        'Amazing product, very useful',
        'Worst experience ever',
        'It is okay, not bad',
        'Loved it, excellent quality',
        'Very disappointing'
    ]
}

df_sent = pd.DataFrame(data)

def sentiment(text):
    score = TextBlob(text).sentiment.polarity
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df_sent['Sentiment'] = df_sent['review'].apply(sentiment)
df_sent
```

```python
df_sent['Sentiment'].value_counts().plot(kind='bar')
plt.title("Sentiment Analysis Result")
plt.show()
```
